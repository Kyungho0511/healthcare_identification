```python
import pandas as pd
```

```python
file_path_2010 = '../data/S2701 Selected Characteristics of Health Insurance Coverage in the United States/ACSST1Y2010.S2701-Data.csv'
file_path_2022 = '../data/S2701 Selected Characteristics of Health Insurance Coverage in the United States/ACSST1Y2022.S2701-Data.csv'
df_2010 = pd.read_csv(file_path_2010)
df_2022 = pd.read_csv(file_path_2022)
```

```python
# Filter columns by specific index positions
filtered_df_2010 = df_2010.iloc[:, [0, 1, 2, 4, 6]]
filtered_df_2022 = df_2022.iloc[:, [0, 1, 2, 124, 246]]

# Extract last two characters of string values in the first column
filtered_df_2010['GEO_ID'] = filtered_df_2010['GEO_ID'].str[-2:]
filtered_df_2022['GEO_ID'] = filtered_df_2022['GEO_ID'].str[-2:]

# Change column names
column_names = ['GEO_ID', 'STATE', 'S2701_POPULATION', 'S2701_UNINSURED_NUMBER', 'S2701_UNINSURED_PERCENT']
filtered_df_2010.columns = column_names
filtered_df_2022.columns = column_names

# Convert all values to float
filtered_df_2010 = filtered_df_2010.drop(0)
filtered_df_2022 = filtered_df_2022.drop(0)
columns_to_convert = filtered_df_2010.columns[-3:]
for column in columns_to_convert:
    filtered_df_2010[column] = filtered_df_2010[column].astype(float)
    filtered_df_2022[column] = filtered_df_2022[column].astype(float)

# Change the insured percent to the uninsured for filtered_df_2022 Data
filtered_df_2022['S2701_UNINSURED_PERCENT'] = 100 - filtered_df_2022['S2701_UNINSURED_PERCENT']
filtered_df_2022['S2701_UNINSURED_PERCENT'] = filtered_df_2022['S2701_UNINSURED_PERCENT'].round(1)

print(filtered_df_2022)
```

```python
# File path where you want to save the CSV file
output_file_path_2010 = '../data/outputs/insurance_coverage_by_states_2010.csv'
output_file_path_2022 = '../data/outputs/insurance_coverage_by_states_2022.csv'

# Export 'result_df' DataFrame to a CSV file
filtered_df_2010.to_csv(output_file_path_2010, index=False)
filtered_df_2022.to_csv(output_file_path_2022, index=False)
```
